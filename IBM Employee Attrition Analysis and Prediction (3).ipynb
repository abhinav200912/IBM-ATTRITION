# Feature Selection

### Filter Method 

```python
#Correlation heatmap helps in checking inter-correlation between columns
```

```python
corrr=Train.corr()
plt.figure(figsize=(25,15))
sns.heatmap(corrr,annot=True)
```

```python
#helps in finding out unnecessary features
def handling_correlation(df,threshold=0.8):
    # creating set to hold the correlated features
    corr_features = set()
    # create the correlation matrix (default to pearson)
    corr_matrix = df.corr()
    #code for selecting all correleated features
    for i in range(len(corr_matrix .columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) >threshold:
                colname = corr_matrix.columns[i]
                corr_features.add(colname)
    return list(corr_features)
#in the time of checking co-relation we have to check with target also that's why I am passing out output_label also
handling_correlation(Train.copy(),0.85)
```

```python
#dropping  features on the basis of filter method
Train.drop('MonthlyIncome',inplace=True,axis=1)
Test.drop('MonthlyIncome',inplace=True,axis=1)
```

# Modeling On Train Set

```python
#Splitting Train Data 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Train.drop(columns = ['Attrition'],axis=1), Train.Attrition)
```

```python
X_train.shape,y_train.shape
```

### 1.Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
model = LogisticRegression()
model.fit(X_train,y_train)
model_predictions = model.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, model_predictions))
print(classification_report(y_test, model_predictions))
```

### 2.Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train)
model_predictions = model.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, model_predictions))
print(classification_report(y_test, model_predictions))
```

### 3.Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
model_predictions = model.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, model_predictions))
print(classification_report(y_test, model_predictions))
```

### 4.K-Neighbors Classification

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train,y_train)
model_predictions = model.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, model_predictions))
print(classification_report(y_test, model_predictions))
```

### 5.Support Vector Classifier

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
model_predictions = model.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, model_predictions))
print(classification_report(y_test, model_predictions))
```

### 6.XGBoost Classifier

```python
from xgboost.sklearn import XGBClassifier
model = XGBClassifier(use_label_encoder=False)
model.fit(X_train,y_train)
model_predictions = model.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, model_predictions))
print(classification_report(y_test, model_predictions))
```

```python
#tensorflow
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout

#EarlyStopping
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#building sequential neural network
model = Sequential()

model.add(Dense(units=100,activation='relu'))
model.add(Dropout(0.5)) 

model.add(Dense(units=50,activation='relu'))
model.add(Dropout(0.5)) 

model.add(Dense(units=25,activation='relu'))
model.add(Dropout(0.5)) 

model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer = 'adam')
```

```python
#using early stopping to prevent overfitting
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1, patience=5)
```

```python
#fitting model into training datasets
model.fit(x = X_train,
          y = y_train,
          batch_size = 128,
          epochs = 100,
          validation_data = (X_test,y_test),verbose =1,
          callbacks = [early_stop]
          )
```

```python
#plotting training and validation losses
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
from sklearn.metrics import confusion_matrix
#confusion matrix
y_pred = model.predict_classes(X_test)
print(confusion_matrix(y_test,y_pred))
```
